```python
from pynq.overlays.base import BaseOverlay
base = BaseOverlay("base.bit")
from pynq.lib.video import *
import cv2
```

```python
def cap_image():
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("camera not opened")
        return
    print("camera opened")
    
    ret, frame = cap.read()
    
    if not ret:
        print("not read")
        cap.release()
        return
    filename = 'test.jpg'
    cv2.imwrite(filename, frame)
    
    cap.release()
```

```python
cap_image()
```
